```python
import pandas as pd

dm = pd.read_json('messages.json')
```

```python
hasil = []
for pesan in dm['conversation']:
    for item in pesan[1:]:
        hasil.append(item)
```

```python
isi_dm = pd.DataFrame(hasil)
isi_dm.head()
```

```python
isi_dm.columns
```

```python
df = isi_dm.copy()
df.drop(['action', 'animated_media_images', 'heart', 'is_random',
       'likes', 'link', 'media', 'media_owner', 'media_share_caption',
       'media_share_url', 'media_url', 'mentioned_username',
       'profile_share_name', 'profile_share_username', 'story_share',
       'user', 'video_call_action'], axis=1, inplace=True)
```
